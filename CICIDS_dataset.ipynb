```
import matplotlib.pyplot as plt
import seaborn as sns


target_counts = df['target'].value_counts().reset_index()
target_counts.columns = ['target', 'count']

# Create the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='target', y='count', hue='target', data=target_counts, palette='viridis', legend=False)
plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```
missing_rows = df[df.isnull().any(axis=1)]
display(missing_rows)
```

```
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```
import matplotlib.pyplot as plt


target_proportions = df['target'].value_counts(normalize=True)


plt.figure(figsize=(8, 8))
plt.pie(target_proportions, labels=target_proportions.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis.colors)
plt.title('Proportion of Target Variable')
plt.axis('equal')
plt.show()
```

```
plt.figure(figsize=(10, 6))
plt.scatter(df[' Flow Duration'], df[' Total Fwd Packets'], alpha=0.5)
plt.title('Flow Duration vs. Total Fwd Packets')
plt.xlabel('Flow Duration')
plt.ylabel('Total Fwd Packets')
plt.grid(True)
plt.show()
```
